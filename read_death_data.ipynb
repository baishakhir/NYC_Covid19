```python
import xlrd
from datetime import datetime
import os
import sys
import csv

DirPath='/Users/bray/Covid19-nyc/Deaths/'
Boroughs = ['Bronx','Brooklyn','Manhattan','Queens', 'Staten Island']

def get_date(loc):
    #covid-19-daily-data-summary-deaths-04032020-2
    date_string = loc.split('deaths-')[1].split('-')[0]
    datetime_object = datetime.strptime(date_string, '%m%d%Y')
    return datetime_object.date()
    
    

def death_per_borough(input_file, output_file, date=None):
    
    if(date is None):
        date=get_date(input_file)
    
    #print(date)
    loc=os.path.join(DirPath, input_file)
    #print(loc)
    #sys.exit(0)
    
    wb = xlrd.open_workbook(loc) 
    sheet = wb.sheet_by_index(0) 
    
    for i in range(sheet.nrows): 
        
        first_column = sheet.cell_value(i,0)
        if any(borough in first_column for borough in Boroughs):
            borough = sheet.cell_value(i,0).split('-')[1].strip()
            
            row_values = sheet.row_values(i)
            total_death = death_data = sheet.cell_value(i,sheet.ncols-2) 
           
            if '(' in str(total_death):
                total_death = death_data.split('(')[0].strip()
            
            x = ",".join((str(date),borough,str(total_death)))
            print(x)
            print(x, file=output_file)
           
     
def read_from_all_excel():
    
    file_name=os.path.join(DirPath, 'death_per_borough.csv')
    
    output_file = open(file_name, 'w') 
    print('Date,Borough,Death', file=output_file)
    output_file.close()
    #death_per_borough
    
    output_file = open(file_name, 'a+') 
    death_per_borough("covid-19-daily-data-summary-deaths-03222020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-03232020-2.xlsx", output_file) 
    death_per_borough("covid-19-daily-data-summary-deaths-03242020-2.xlsx", output_file) 
    death_per_borough("covid-19-daily-data-summary-deaths-03252020-2.xlsx", output_file) 
    death_per_borough("covid-19-daily-data-summary-deaths-03262020-2.xlsx", output_file) 
    death_per_borough("covid-19-daily-data-summary-deaths-03272020-2.xlsx", output_file) 
    death_per_borough("covid-19-daily-data-summary-deaths-03282020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-03292020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-03302020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-03312020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-04012020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-04022020-2.xlsx", output_file)
    death_per_borough("covid-19-daily-data-summary-deaths-04032020-2.xlsx", output_file)
    
    output_file.close()
```
